```python
import MetaTrader5 as mt5

# Verbinden met de MT5 terminal
if not mt5.initialize():
    print("MT5 initialisatie mislukt")
    quit()

# Ophalen van de beschikbare symbolen
symbols = mt5.symbols_get()

# Lijst van symbolen weergeven
for symbol in symbols:
    print(symbol.name)

# Verbinding sluiten
mt5.shutdown()
```
