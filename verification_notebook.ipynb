```python
# Sophia Framework Verificatie
# --------------------------
# Deze notebook test en demonstreert de functionaliteit van het Sophia Trading Framework

import os
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import time
from datetime import datetime, timedelta
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Zorg dat we de Sophia modules kunnen importeren
script_dir = os.path.dirname(os.path.abspath('__file__'))
project_root = os.path.dirname(script_dir)  # Als deze notebook in de hoofdmap staat
if project_root not in sys.path:
    sys.path.insert(0, project_root)

# Importeer Sophia componenten
from src.connector import MT5Connector
from src.risk import RiskManager
from src.strategy import TurtleStrategy
from src.utils import load_config

# Stel grafiekstijl in
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("viridis")
plt.rcParams['figure.figsize'] = (15, 8)
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10

# %% [markdown]
# ## 1. Systeemverificatie
#
# Laten we eerst controleren of alle componenten correct zijn geïnstalleerd en of we toegang hebben tot de juiste modules.

# %%
# Controleer beschikbare modules
import MetaTrader5 as mt5
print(f"MetaTrader5 versie: {mt5.__version__}")
print(f"Python versie: {sys.version}")
print(f"Pandas versie: {pd.__version__}")
print(f"NumPy versie: {np.__version__}")

# %% [markdown]
# ## 2. MT5 Connectiviteit
#
# We testen of de MT5Connector correct verbinding kan maken met het MetaTrader 5 platform.

# %%
# Laad configuratie
try:
    # Directe padverwijzing voor maximale betrouwbaarheid
    config_path = r"C:\Users\Gebruiker\PycharmProjects\Sophia\config\settings.json"
    config = load_config(config_path)

    if config:
        print("✅ Configuratie succesvol geladen vanaf absoluut pad")
    else:
        print("⚠️ Configuratie niet gevonden op absoluut pad, fallback gebruiken...")
        # Fallback als vangnet
        fallback_path = os.path.join(project_root, "config", "settings.json")
        config = load_config(fallback_path)

    if not config:
        print("❌ Fout bij laden configuratie, gebruik standaardwaarden")
        # Standaard test configuratie
        config = {
            "mt5": {
                "server": "FTMO-Demo2",
                "login": 1520533067,
                "password": "UP7d??y4Wg",
                "mt5_path": "C:\\Program Files\\FTMO Global Markets MT5 Terminal\\terminal64.exe",
            },
            "symbols": ["EURUSD", "USDJPY", "GBPUSD", "AUDUSD"],
            "timeframe": "H4",
            "interval": 300,
            "risk": {"risk_per_trade": 0.01, "max_daily_loss": 0.05},
        }
    else:
        print("✅ Configuratie succesvol geladen")

    print("MetaTrader 5 configuratie:")
    print(f"  Server: {config['mt5'].get('server')}")
    print(f"  Login: {config['mt5'].get('login')}")
    print(f"  Pad: {config['mt5'].get('mt5_path')}")

except Exception as e:
    print(f"❌ Fout bij configuratiecontrole: {e}")
    # Zorg ervoor dat we een standaardconfiguratie hebben
    config = {
        "mt5": {
            "server": "FTMO-Demo2",
            "login": 1520533067,
            "password": "UP7d??y4Wg",
            "mt5_path": "C:\\Program Files\\FTMO Global Markets MT5 Terminal\\terminal64.exe",
        },
        "symbols": ["EURUSD", "USDJPY"],
        "timeframe": "H4",
        "interval": 300,
        "risk": {"risk_per_trade": 0.01, "max_daily_loss": 0.05},
    }

# %%
# Maak verbinding met MT5
mt5_connected = False
connector = None
account_info = {}

try:
    # Initialize connector
    connector = MT5Connector(config.get("mt5", {}))

    # Maak verbinding
    if connector.connect():
        mt5_connected = True
        print("✅ Verbinding met MT5 succesvol")

        # Haal account informatie op
        account_info = connector.get_account_info()
        print(f"\nAccount informatie:")
        print(f"  Balans: {account_info.get('balance', 'onbekend')} {account_info.get('currency', '')}")
        print(f"  Equity: {account_info.get('equity', 'onbekend')} {account_info.get('currency', '')}")
        print(f"  Margin: {account_info.get('margin', 'onbekend')} {account_info.get('currency', '')}")
        print(f"  Free margin: {account_info.get('free_margin', 'onbekend')} {account_info.get('currency', '')}")

        # Haal beschikbare symbolen op
        symbols = connector.get_available_symbols()
        if symbols:
            print(f"\nAantal beschikbare symbolen: {len(symbols)}")
            print(f"Eerste 10 symbolen: {symbols[:10]}")
        else:
            print("\n❌ Geen symbolen gevonden")
    else:
        print("❌ Verbinding met MT5 mislukt")

except Exception as e:
    print(f"❌ Fout bij MT5 verbinding: {e}")

# %% [markdown]
# ## 3. Markturen Analyse
#
# Nu gaan we controleren of de Sophia trading engine de juiste markturen kan detecteren. Dit is cruciaal voor het juist plannen van trades.

# %%
def markturen_analyse(symbol="EURUSD"):
    """Analyseer markturen voor een specifiek symbool"""
    # Haal historische data op om markturen te bepalen
    # We kijken naar 1-minuut data voor 2 weken
    end_time = datetime.now()
    start_time = end_time - timedelta(days=14)

    # Ophalen minutendata voor markturen detectie
    try:
        # Eerst controleren of we niet al MT5 data voor dit symbool hebben opgehaald
        if not hasattr(mt5, 'last_data') or mt5.last_symbol != symbol:
            rates = mt5.copy_rates_range(symbol, mt5.TIMEFRAME_M1,
                                         start_time, end_time)
            mt5.last_data = rates
            mt5.last_symbol = symbol
        else:
            rates = mt5.last_data

        if rates is None or len(rates) == 0:
            return None, None

        # Converteer naar DataFrame
        df_min = pd.DataFrame(rates)
        df_min['time'] = pd.to_datetime(df_min['time'], unit='s')

        # Voeg dag van de week toe
        df_min['day_of_week'] = df_min['time'].dt.day_name()
        df_min['hour'] = df_min['time'].dt.hour
        df_min['minute'] = df_min['time'].dt.minute

        # Groepeer per dag en uur
        time_counts = df_min.groupby(['day_of_week', 'hour']).size().reset_index(name='count')

        # Bepaal welke uren de markt open is
        pivot_table = time_counts.pivot(index='hour', columns='day_of_week', values='count')

        return df_min, pivot_table

    except Exception as e:
        print(f"Fout bij markturen analyse: {e}")
        return None, None

# %%
# Initialiseer variabelen om later op te controleren
df_min = None
pivot_table = None

# Test alleen als MT5 verbinding is geslaagd
if mt5_connected and connector:
    symbol = config['symbols'][0]  # Bijvoorbeeld "EURUSD"
    df_min, pivot_table = markturen_analyse(symbol)

    if df_min is not None and pivot_table is not None:
        print(f"Markturen analyse voor {symbol}")

        # Teken heatmap van markturen
        plt.figure(figsize=(15, 10))
        sns.heatmap(pivot_table, cmap="YlGnBu", annot=False, cbar_kws={'label': 'Aantal minuten data'})
        plt.title(f'Markturen voor {symbol} (Licht = Open, Donker = Gesloten of Dunne Markt)')
        plt.ylabel('Uur van de dag (UTC)')
        plt.xlabel('Dag van de week')
        plt.tight_layout()
        plt.show()

        # Berekenen en weergeven van samenvatting
        weekdays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
        active_hours = pivot_table.fillna(0).astype(bool).sum(axis=1)
        active_days = pivot_table.fillna(0).astype(bool).sum(axis=0)

        print("\nActieve handelsdagen:")
        for day, count in zip(active_days.index, active_days.values):
            status = "✅ Open" if count > 12 else "❌ Gesloten" if count == 0 else "⚠️ Gedeeltelijk"
            print(f"  {day}: {status} ({count} actieve uren)")

        print("\nMarkturen per dag (UTC):")
        for day in weekdays:
            if day in pivot_table.columns:
                active = pivot_table[day].fillna(0) > 0
                if active.any():
                    open_hours = active.index[active].tolist()
                    if len(open_hours) > 0:
                        start = min(open_hours)
                        end = max(open_hours) + 1
                        print(f"  {day}: {start:02d}:00 - {end:02d}:00 UTC")
                    else:
                        print(f"  {day}: Gesloten")
                else:
                    print(f"  {day}: Gesloten")
    else:
        print(f"Kan geen data ophalen voor {symbol}")
else:
    print("Markturen analyse overgeslagen - Geen MT5 verbinding")

# %% [markdown]
# ## 4. Historische Data Kwaliteit
#
# Nu gaan we de kwaliteit van historische data controleren die Sophia ophaalt.

# %%
def test_historische_data(connector, symbol="EURUSD", timeframe="H4", bars=100):
    """Test het ophalen van historische data voor een symbool"""
    try:
        # Haal historische data op
        data = connector.get_historical_data(symbol, timeframe, bars)

        if data is None:
            print(f"❌ Geen data ontvangen voor {symbol}, {timeframe}")
            return None

        print(f"✅ Data ontvangen voor {symbol}, {timeframe}")
        print(f"  Aantal bars: {len(data)}")
        print(f"  Eerste datum: {data['time'].min()}")
        print(f"  Laatste datum: {data['time'].max()}")
        print(f"  Tijdspanne: {(data['time'].max() - data['time'].min()).days} dagen")

        # Data kwaliteitscontrole
        missing_values = data.isnull().sum().sum()
        if missing_values > 0:
            print(f"⚠️ {missing_values} ontbrekende waarden gedetecteerd")
        else:
            print("✅ Geen ontbrekende waarden")

        # Controleer op duplicaten
        duplicates = data.duplicated(subset=['time']).sum()
        if duplicates > 0:
            print(f"⚠️ {duplicates} dubbele tijdstempels gedetecteerd")
        else:
            print("✅ Geen dubbele tijdstempels")

        # Consistentiecontrole voor OHLC data (High >= Low, etc.)
        inconsistent = ((data['high'] < data['low']) |
                        (data['high'] < data['open']) |
                        (data['high'] < data['close']) |
                        (data['low'] > data['open']) |
                        (data['low'] > data['close'])).sum()

        if inconsistent > 0:
            print(f"⚠️ {inconsistent} inconsistente OHLC records")
        else:
            print("✅ OHLC consistentie geverifieerd")

        return data

    except Exception as e:
        print(f"❌ Fout bij historische data test: {e}")
        return None

# %%
# Initialiseer variabelen om later op te controleren
history_test_results = {}

# Test alleen als MT5 verbinding is geslaagd
if mt5_connected and connector:
    # Test meerdere symbolen en timeframes
    timeframes = ["H1", "H4", "D1"]
    symbols = config.get('symbols', ["EURUSD", "USDJPY"])

    for symbol in symbols[:2]:  # Beperk tot 2 symbolen voor de notebook
        symbol_results = {}
        print(f"\n=== Historische data test voor {symbol} ===")

        for tf in timeframes:
            print(f"\nTimeframe: {tf}")
            data = test_historische_data(connector, symbol, tf, 200)
            symbol_results[tf] = data

        history_test_results[symbol] = symbol_results
else:
    print("Historische data tests overgeslagen - Geen MT5 verbinding")

# %%
# Visualiseer de resultaten voor één symbool (OHLC-grafiek)
if history_test_results and len(history_test_results) > 0:
    symbol = list(history_test_results.keys())[0]
    tf = "D1"  # Dagelijks timeframe

    if symbol in history_test_results and tf in history_test_results[symbol] and history_test_results[symbol][tf] is not None:
        data = history_test_results[symbol][tf]

        # Plot OHLC data
        plt.figure(figsize=(16, 8))

        # Maak een candlestick chart
        # We beperken tot de laatste 50 bars voor betere visualisatie
        last_n_bars = 50
        subset = data.iloc[-last_n_bars:].copy()

        width = 0.6
        width2 = 0.1

        # Genereer x-as waarden
        subset['x'] = range(len(subset))

        # Vul kleurcodes voor candlesticks
        subset['colors'] = ['red' if close < open else 'green' for close, open in zip(subset['close'], subset['open'])]

        # Teken candlesticks
        plt.bar(subset['x'], subset['high'] - subset['low'], width=width2, bottom=subset['low'], color='black', alpha=0.8)
        plt.bar(subset['x'], subset['close'] - subset['open'], width=width, bottom=subset['open'], color=subset['colors'], alpha=0.8)

        # Voeg x-as labels toe
        tick_positions = np.linspace(0, len(subset) - 1, 10, dtype=int)
        plt.xticks(tick_positions, [subset.iloc[pos]['time'].strftime('%Y-%m-%d') for pos in tick_positions], rotation=45)

        plt.title(f'{symbol} - {tf} Candlestick Chart')
        plt.xlabel('Datum')
        plt.ylabel('Prijs')
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()

        # Controleer ook voor gaps in de data
        subset = data.copy()

        if tf == "D1":
            # Verwachte frequentie is dagelijks (werkdagen)
            expected_index = pd.date_range(start=subset['time'].min(), end=subset['time'].max(), freq='B')
            actual_dates = set(subset['time'].dt.date)
            missing_dates = [date for date in expected_index if date.date() not in actual_dates and date.weekday() < 5]

            if missing_dates:
                print(f"\n⚠️ Ontbrekende handelsdagen gedetecteerd ({len(missing_dates)}):")
                for date in missing_dates[:5]:  # Toon max 5 ontbrekende data
                    print(f"  - {date.date()}")
                if len(missing_dates) > 5:
                    print(f"  - ... en {len(missing_dates) - 5} meer")
            else:
                print("\n✅ Alle verwachte handelsdagen zijn aanwezig in de data")
    else:
        print(f"Geen data beschikbaar voor {symbol} {tf}")
else:
    print("Visualisatie overgeslagen - Geen historische data resultaten")

# %% [markdown]
# ## 5. Strategy Backtesting Verificatie
#
# Nu gaan we controleren of het Sophia framework signalen correct kan genereren op basis van historische data.

# %%
# Initialiseer de strategy en risk manager
strategy_test_results = {}

if mt5_connected and connector:
    risk_manager = RiskManager(config.get("risk", {}))
    strategy = TurtleStrategy(connector, risk_manager, config.get("strategy", {
        "entry_period": 20,
        "exit_period": 10,
        "atr_period": 14,
        "vol_filter": True
    }))

    def test_strategie(strategy, symbol="EURUSD", timeframe="H4", bars=200):
        """Test of de strategie correcte signalen genereert"""
        try:
            # Haal historische data op
            data = connector.get_historical_data(symbol, timeframe, bars)

            if data is None or len(data) < 50:
                print(f"❌ Onvoldoende data voor {symbol}")
                return None, [], [], [], []

            # Bereken indicators
            data_with_indicators = strategy.calculate_indicators(data)

            # Check signalen voor elke bar
            signals = []
            signal_dates = []
            signal_prices = []
            signal_types = []

            for i in range(30, len(data_with_indicators)):
                # Maak een subset van data tot deze bar
                historical_subset = data_with_indicators.iloc[:i+1].copy()

                # Controleer signaal
                result = strategy.check_signals_with_data(symbol, historical_subset)

                if result.get("signal"):
                    signals.append(result.get("signal"))
                    signal_dates.append(historical_subset.iloc[-1]['time'])
                    signal_prices.append(historical_subset.iloc[-1]['close'])
                    signal_types.append(result.get("meta", {}).get("reason", "unknown"))

            print(f"\nStrategie test voor {symbol} {timeframe}")
            print(f"  Verwerkte bars: {len(data_with_indicators)}")
            print(f"  Signalen gedetecteerd: {len(signals)}")
            print(f"  - BUY signalen: {signals.count('BUY')}")
            print(f"  - SELL signalen: {signals.count('SELL')}")
            print(f"  - EXIT signalen: {signals.count('CLOSE_BUY') + signals.count('CLOSE_SELL')}")

            return data_with_indicators, signals, signal_dates, signal_prices, signal_types

        except Exception as e:
            print(f"❌ Fout bij strategie test: {e}")
            import traceback
            traceback.print_exc()
            return None, [], [], [], []

    # Test de strategie op verschillende symbolen
    for symbol in symbols[:1]:  # Beperk tot 1 symbool voor notebook
        print(f"\n=== Strategie test voor {symbol} ===")
        result = test_strategie(strategy, symbol, "H4", 500)
        strategy_test_results[symbol] = result
else:
    print("Strategie tests overgeslagen - Geen MT5 verbinding")

# %%
# Visualiseer de strategie resultaten
if strategy_test_results and len(strategy_test_results) > 0:
    symbol = list(strategy_test_results.keys())[0]

    if symbol in strategy_test_results and strategy_test_results[symbol][0] is not None:
        data, signals, signal_dates, signal_prices, signal_types = strategy_test_results[symbol]

        # Beperk tot de laatste 100 bars voor betere visualisatie
        last_n_bars = 100
        subset = data.iloc[-last_n_bars:].copy()

        plt.figure(figsize=(16, 10))

        # Plot prijsdata
        plt.plot(subset['time'], subset['close'], 'k-', label='Sluitkoers', alpha=0.7)

        # Plot Donchian channels
        plt.plot(subset['time'], subset['entry_high'], 'g--', label='Entry High', alpha=0.5)
        plt.plot(subset['time'], subset['entry_low'], 'r--', label='Entry Low', alpha=0.5)
        plt.plot(subset['time'], subset['exit_high'], 'g:', label='Exit High', alpha=0.4)
        plt.plot(subset['time'], subset['exit_low'], 'r:', label='Exit Low', alpha=0.4)

        # Filter signalen voor deze weergaveperiode
        start_date = subset['time'].min()
        end_date = subset['time'].max()
        filtered_signals = [(s, d, p, t) for s, d, p, t in zip(signals, signal_dates, signal_prices, signal_types)
                             if start_date <= d <= end_date]

        # Plot signalen
        for signal, date, price, reason in filtered_signals:
            if signal == 'BUY':
                plt.scatter(date, price, marker='^', color='green', s=120, label='BUY' if 'BUY' not in plt.gca().get_legend_handles_labels()[1] else "")
            elif signal == 'SELL':
                plt.scatter(date, price, marker='v', color='red', s=120, label='SELL' if 'SELL' not in plt.gca().get_legend_handles_labels()[1] else "")
            elif signal == 'CLOSE_BUY':
                plt.scatter(date, price, marker='o', color='blue', s=100, label='CLOSE_BUY' if 'CLOSE_BUY' not in plt.gca().get_legend_handles_labels()[1] else "")
            elif signal == 'CLOSE_SELL':
                plt.scatter(date, price, marker='o', color='purple', s=100, label='CLOSE_SELL' if 'CLOSE_SELL' not in plt.gca().get_legend_handles_labels()[1] else "")

        plt.title(f'{symbol} - Strategie Signalen (Laatste {last_n_bars} bars)')
        plt.xlabel('Datum')
        plt.ylabel('Prijs')
        plt.grid(True, alpha=0.3)
        plt.legend(loc='best')
        plt.tight_layout()
        plt.show()

        # Toon signaal details in een tabel
        if filtered_signals:
            print("\nGedetecteerde signalen in weergaveperiode:")
            for signal, date, price, reason in filtered_signals:
                print(f"  {date.strftime('%Y-%m-%d %H:%M')}: {signal} @ {price:.5f} - Reden: {reason}")
        else:
            print("\nGeen signalen gedetecteerd in weergaveperiode")
    else:
        print(f"Geen strategie resultaten beschikbaar voor {symbol}")
else:
    print("Visualisatie overgeslagen - Geen strategie resultaten")

# %% [markdown]
# ## 6. Risk Management Verificatie
#
# Nu gaan we het risk management systeem testen om te zien of het de correcte positiegroottes berekent.

# %%
# Initialiseer variabelen om later op te controleren
risk_results = None

if mt5_connected:
    def test_risk_management(risk_manager, account_balances=[10000, 50000, 100000]):
        """Test risk management met verschillende scenario's"""

        stop_ranges = {
            "zeer_nauw": 5,   # 5 pips
            "nauw": 20,       # 20 pips
            "middel": 50,     # 50 pips
            "wijd": 100,      # 100 pips
        }

        results = {}
        for balance in account_balances:
            balance_results = {}

            for stop_name, pips in stop_ranges.items():
                # Bereken positiegrootte voor verschillende symbolen
                symbol_results = {}

                for symbol in ["EURUSD", "USDJPY", "GBPUSD"]:
                    # Bepaal correcte pip multiplier (JPY vs andere)
                    pip_multiplier = 0.01 if symbol.startswith("JPY") else 0.0001
                    stop_loss_distance = pips * pip_multiplier

                    # Test met verschillende entry prijzen
                    entry_prices = {
                        "EURUSD": 1.1000,
                        "USDJPY": 150.00,
                        "GBPUSD": 1.2500
                    }
                    entry_price = entry_prices.get(symbol, 1.0)
                    stop_loss = entry_price - stop_loss_distance  # Altijd onder de entry prijs voor dit voorbeeld

                    # Bereken positiegrootte
                    position_size = risk_manager.calculate_position_size(
                        balance, entry_price, stop_loss, symbol
                    )

                    # Risico in account valuta
                    risk_amount = (entry_price - stop_loss) * position_size * (10000 if pip_multiplier == 0.0001 else 1000)
                    risk_percentage = (risk_amount / balance) * 100

                    symbol_results[symbol] = {
                        "position_size": position_size,
                        "risk_amount": risk_amount,
                        "risk_percentage": risk_percentage,
                        "entry_price": entry_price,
                        "stop_loss": stop_loss,
                        "pips_at_risk": pips
                    }

                balance_results[stop_name] = symbol_results
            results[balance] = balance_results

        return results

    # Test risk management
    risk_results = test_risk_management(risk_manager)

    # Visualiseer de resultaten in een bar chart
    if risk_results:
        balances = list(risk_results.keys())
        stop_types = list(risk_results[balances[0]].keys())
        symbols = list(risk_results[balances[0]][stop_types[0]].keys())

        # Kies één balans voor visualisatie
        balance = balances[0]

        # Plot positiegroottes per stop-afstand
        plt.figure(figsize=(15, 8))
        bar_width = 0.2
        index = np.arange(len(stop_types))

        for i, symbol in enumerate(symbols):
            sizes = [risk_results[balance][stop][symbol]["position_size"] for stop in stop_types]
            plt.bar(index + i*bar_width, sizes, bar_width,
                    alpha=0.8,
                    label=symbol)

        plt.xlabel('Stop Loss Afstand')
        plt.ylabel('Positiegrootte (lots)')
        plt.title(f'Positiegrootte vs. Stop Loss Afstand (Account Balans: ${balance})')
        plt.xticks(index + bar_width, stop_types)
        plt.legend()
        plt.grid(axis='y', alpha=0.3)
        plt.tight_layout()
        plt.show()

        # Plot risicopercentage per stop-afstand
        plt.figure(figsize=(15, 8))

        for i, symbol in enumerate(symbols):
            percentages = [risk_results[balance][stop][symbol]["risk_percentage"] for stop in stop_types]
            plt.bar(index + i*bar_width, percentages, bar_width,
                    alpha=0.8,
                    label=symbol)

        plt.xlabel('Stop Loss Afstand')
        plt.ylabel('Risico Percentage (%)')
        plt.title(f'Risico Percentage vs. Stop Loss Afstand (Account Balans: ${balance})')
        plt.xticks(index + bar_width, stop_types)
        plt.axhline(y=1.0, color='r', linestyle='--', label='1% Risico Target')
        plt.legend()
        plt.grid(axis='y', alpha=0.3)
        plt.tight_layout()
        plt.show()

        # Print een tabel met gedetailleerde resultaten
        print(f"Gedetailleerde Risk Management Resultaten (Account Balans: ${balance}):\n")
        print(f"{'Stop Type':<12} {'Symbol':<10} {'Lot Size':<10} {'Risk Amt':<10} {'Risk %':<10} {'Entry':<10} {'Stop':<10} {'Pips':<8}")
        print("-" * 80)

        for stop_type in stop_types:
            for symbol in symbols:
                result = risk_results[balance][stop_type][symbol]
                print(f"{stop_type:<12} {symbol:<10} {result['position_size']:<10.2f} "
                    f"${result['risk_amount']:<9.2f} {result['risk_percentage']:<9.2f}% "
                    f"{result['entry_price']:<9.4f} {result['stop_loss']:<9.4f} {result['pips_at_risk']:<8.0f}")
else:
    print("Risk management tests overgeslagen - Geen MT5 verbinding")

# %% [markdown]
# ## 7. Trade Execution Simulatie
#
# Laten we nu een volledige handelscyclus simuleren om te verifiëren dat Sophia orders correct kan plaatsen en bijhouden.
# (Voor een notebook houden we dit in de demo/simulatie modus)

# %%
# Initialiseer variabelen om later op te controleren
simulatie_succesvol = False

if mt5_connected and connector and 'strategy' in locals():
    def simuleer_handelsoperatie(strategy, connector, symbol="EURUSD", demo_mode=True):
        """Simuleer een volledige handelsoperatie (entry en exit)"""
        print(f"\n=== Handelsoperatie Simulatie voor {symbol} ===")

        if demo_mode:
            print("⚠️ Draait in DEMO MODUS - Echte orders worden niet geplaatst")

            # Simuleer data ophalen
            data = connector.get_historical_data(symbol, "H4", 100)
            if data is None or len(data) < 30:
                print("❌ Onvoldoende data voor simulatie")
                return False

            # Simuleer entry signaal
            entry_price = data['close'].iloc[-1]
            stop_loss = entry_price * 0.995  # 0.5% onder entry prijs
            account_balance = 10000.0

            # Simuleer positiegrootte berekening
            position_size = strategy.risk_manager.calculate_position_size(
                account_balance, entry_price, stop_loss, symbol
            )

            print(f"\n✅ SIMULATIE: Entry Signaal voor {symbol}")
            print(f"  Entry Prijs: {entry_price:.5f}")
            print(f"  Stop Loss: {stop_loss:.5f}")
            print(f"  Positiegrootte: {position_size:.2f} lots")

            # Simuleer het plaatsen van de order
            print(f"\n✅ SIMULATIE: Order Geplaatst")
            print(f"  Order Type: BUY")
            print(f"  Symbol: {symbol}")
            print(f"  Volume: {position_size}")
            print(f"  Prijs: {entry_price:.5f}")
            print(f"  Stop Loss: {stop_loss:.5f}")
            print(f"  Take Profit: {entry_price * 1.01:.5f}")  # 1% boven entry prijs

            # Simuleer het bijhouden van de positie
            strategy.positions[symbol] = {
                "direction": "BUY",
                "entry_price": entry_price,
                "stop_loss": stop_loss,
                "size": position_size,
                "entry_time": datetime.now(),
                "order_id": "12345"  # Demo order ID
            }

            print(f"\n✅ SIMULATIE: Positie bijgewerkt in strategie tracking")
            print(f"  Huidige posities: {list(strategy.positions.keys())}")

            # Simuleer de positie te volgen gedurende tijd
            print(f"\n✅ SIMULATIE: Positie monitoring")
            print(f"  Huidige prijs: {entry_price * 1.002:.5f}")  # Huidige prijs 0.2% boven entry
            print(f"  Unrealized P&L: ${position_size * (entry_price * 1.002 - entry_price) * 10000:.2f}")

            # Simuleer een exit signaal
            print(f"\n✅ SIMULATIE: Exit Signaal voor {symbol}")
            exit_price = entry_price * 1.008  # 0.8% winst

            # Simuleer het sluiten van de positie
            print(f"\n✅ SIMULATIE: Positie Gesloten")
            print(f"  Exit Prijs: {exit_price:.5f}")
            print(f"  P&L: ${position_size * (exit_price - entry_price) * 10000:.2f}")

            # Simuleer het verwijderen van de positie uit tracking
            if symbol in strategy.positions:
                del strategy.positions[symbol]

            print(f"\n✅ SIMULATIE: Positie verwijderd uit strategie tracking")
            print(f"  Huidige posities: {list(strategy.positions.keys())}")

            return True
        else:
            # Hier zou de echte implementatie komen als demo_mode=False
            print("❌ Echte ordeplaatsing niet geïmplementeerd in deze notebook")
            return False

    # Simuleer een handelsoperatie
    if len(symbols) > 0:
        simulatie_succesvol = simuleer_handelsoperatie(strategy, connector, symbols[0])

        if simulatie_succesvol:
            print("\n✅ Handelssimulatie succesvol afgerond")

            # Visualiseer de geSimuleerde handelsoperatie
            plt.figure(figsize=(15, 8))

            # Simulatie data
            account_balance = 10000.0
            initial_balance = account_balance

            # Simuleer 10 trades met random resultaten voor visualisatie
            np.random.seed(42)  # Voor reproduceerbaarheid

            # Gebruik echte data voor een realistisch voorbeeld
            data = connector.get_historical_data(symbols[0], "H4", 100)
            if data is not None and len(data) > 30:
                dates = data['time'].iloc[-10:].tolist()

                # Simuleer trades
                balances = [initial_balance]
                trades = []

                for i in range(9):
                    # Simuleer een trade
                    entry_price = data['close'].iloc[-(10-i)]
                    exit_price = data['close'].iloc[-(9-i)]

                    # Winstgevend of verlieslatend?
                    is_profitable = np.random.choice([True, False], p=[0.55, 0.45])  # 55% kans op winst

                    # Bereken resultaat
                    if not is_profitable:
                        # Verlies tussen 0.5% en 1%
                        loss_pct = np.random.uniform(0.5, 1.0)
                        trade_result = -account_balance * (loss_pct / 100)
                    else:
                        # Winst tussen 1% en 3%
                        profit_pct = np.random.uniform(1.0, 3.0)
                        trade_result = account_balance * (profit_pct / 100)

                    # Update balans
                    account_balance += trade_result
                    balances.append(account_balance)

                    # Voeg trade toe aan lijst
                    trades.append({
                        'date': dates[i+1],
                        'type': 'BUY' if i % 2 == 0 else 'SELL',
                        'entry': entry_price,
                        'exit': exit_price,
                        'result': trade_result,
                        'balance': account_balance
                    })

                # Plot equity curve
                plt.subplot(2, 1, 1)
                plt.plot(range(len(balances)), balances, 'b-', linewidth=2)
                plt.title('Gesimuleerde Equity Curve')
                plt.xlabel('Aantal Trades')
                plt.ylabel('Account Balans ($)')
                plt.grid(True, alpha=0.3)

                # Bereken metrics
                total_profit = account_balance - initial_balance
                profit_percentage = (total_profit / initial_balance) * 100
                win_count = sum(1 for trade in trades if trade['result'] > 0)
                loss_count = len(trades) - win_count
                win_rate = (win_count / len(trades)) * 100 if trades else 0

                # Toon metrics als tekstbox
                textbox = (f'Startbalans: ${initial_balance:.2f}\n'
                         f'Eindbalans: ${account_balance:.2f}\n'
                         f'Winst/Verlies: ${total_profit:.2f} ({profit_percentage:.2f}%)\n'
                         f'Winstgevende trades: {win_count}\n'
                         f'Verlieslatende trades: {loss_count}\n'
                         f'Winstpercentage: {win_rate:.2f}%')

                plt.annotate(textbox, xy=(0.02, 0.7), xycoords='axes fraction',
                            bbox=dict(boxstyle="round,pad=0.5", facecolor="white", alpha=0.8))

                # Plot trade resultaten
                plt.subplot(2, 1, 2)
                results = [trade['result'] for trade in trades]
                colors = ['green' if r > 0 else 'red' for r in results]
                plt.bar(range(len(results)), results, color=colors)
                plt.title('Gesimuleerde Trade Resultaten')
                plt.xlabel('Trade Nummer')
                plt.ylabel('Winst/Verlies ($)')
                plt.grid(True, alpha=0.3)
                plt.axhline(y=0, color='black', linestyle='-', linewidth=0.5)

                plt.tight_layout()
                plt.show()

                # Toon trade details
                print("\nGedetailleerde Trade Resultaten:")
                print(f"{'#':<3} {'Type':<6} {'Entry':<10} {'Exit':<10} {'Resultaat':<12} {'Balans':<10}")
                print("-" * 60)
                for i, trade in enumerate(trades, 1):
                    print(f"{i:<3} {trade['type']:<6} {trade['entry']:<10.5f} {trade['exit']:<10.5f} "
                         f"${trade['result']:<11.2f} ${trade['balance']:<10.2f}")
            else:
                print("❌ Onvoldoende data beschikbaar voor visualisatie")
        else:
            print("\n❌ Handelssimulatie mislukt")
    else:
        print("Kan handelssimulatie niet uitvoeren - Geen symbolen beschikbaar")
else:
    print("Handelssimulatie overgeslagen - Geen MT5 verbinding of strategie")

# %% [markdown]
# ## 8. Framework Prestaties Conclusie
#
# Laten we de resultaten samenvatten om te zien hoe goed het Sophia framework presteert.

# %%
# Maak verbinding met MT5 verbreken om resources vrij te geven
if connector:
    connector.disconnect()

print("\n=== Sophia Framework Verificatie Resultaten ===")

# Controleer of alle essentiële tests gelukt zijn
# Gebruiken nu veilige checks met 'in locals()' om TypeError te voorkomen
tests_results = {
    "MT5 Connectiviteit": mt5_connected,
    "Markturen Analyse": 'df_min' in locals() and df_min is not None and 'pivot_table' in locals() and pivot_table is not None,
    "Historische Data": 'history_test_results' in locals() and len(history_test_results) > 0,
    "Strategie Signalen": 'strategy_test_results' in locals() and len(strategy_test_results) > 0,
    "Risk Management": 'risk_results' in locals() and risk_results is not None,
    "Handelssimulatie": 'simulatie_succesvol' in locals() and simulatie_succesvol
}

# Toon resultaten
passed = sum(1 for result in tests_results.values() if result)
total = len(tests_results)
pass_percentage = (passed / total) * 100

print(f"\nSamenvatting van Verificaties: {passed}/{total} succesvol ({pass_percentage:.0f}%)")
print("-" * 60)
for test, result in tests_results.items():
    status = "✅ SUCCES" if result else "❌ MISLUKT"
    print(f"{test:<25}: {status}")

print("\n=== Eindconclusie ===")
if pass_percentage >= 80:
    print("✅ Het Sophia framework werkt correct! Alle essentiële componenten functioneren naar behoren.")
    if pass_percentage < 100:
        print("⚠️ Er zijn enkele aandachtspunten die verder onderzocht kunnen worden.")
elif pass_percentage >= 50:
    print("⚠️ Het Sophia framework functioneert gedeeltelijk. Er zijn verbeteringen nodig.")
else:
    print("❌ Er zijn significante problemen met het Sophia framework die opgelost moeten worden.")

print("\n=== Aanbevelingen ===")
for test, result in tests_results.items():
    if not result:
        if test == "MT5 Connectiviteit":
            print("- Los de MT5 verbindingsproblemen op. Controleer of het platform draait en de inloggegevens correct zijn.")
        elif test == "Markturen Analyse":
            print("- Controleer of de markturen analyse correct werkt. Mogelijk is er een probleem met het ophalen van minutendata.")
        elif test == "Historische Data":
            print("- Controleer de data-ophaalfunctionaliteit. Zorg dat de juiste timeframes en symbolen worden ondersteund.")
        elif test == "Strategie Signalen":
            print("- Optimaliseer de signaaldetectie in de strategie. Controleer of indicators correct worden berekend.")
        elif test == "Risk Management":
            print("- Controleer de risk management implementatie. Zorg dat positiegroottes correct worden berekend.")
        elif test == "Handelssimulatie":
            print("- Controleer de handelslogica. Zorg dat orders correct worden opgebouwd en bijgehouden.")

print("\nBedankt voor het gebruiken van de Sophia Framework Verificatie Notebook!")
```
